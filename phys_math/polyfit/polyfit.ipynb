# The effect of a shift

```python
x = np.linspace(0, 10)
pivot = 10
a, b, c, s = 2, 4, 5, 10
y = a + b*(x - s) + c*(x - s)**2

def compute(x, params):
    res = np.zeros_like(x)
    for i, p in enumerate(params):
        res += p * x**i
    return res
```

```python
params = np.polyfit(x, y, 2)[::-1]
print(params)
```

```python
_, ax = plt.subplots()
ax.scatter(x, y)

ax.plot(x, compute(x, params))
```

### We fit it perfectly. How?

We start with 

$y = a + b (x - s) + c(x - s)^2$

And we fit,

$y = a' + b'x + c'x$

Expanding out the first equation, we get

$y = (a - bs + s^2) + x(b - 2cs) + cx^2$

And so,

$ a' = a - bs + cs^2 $, $b' = b - 2cs$ and $c' = c$
