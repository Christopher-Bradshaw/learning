# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

# Note that this has a ton of MPL config!
%run ~/.jupyter/config.ipy
```

```python
x = np.linspace(0, 2*np.pi)
y_sin = np.sin(x)
y_cos = np.cos(x)
```

# The basic plot

```python
# There are many ways to create a figure and add axes, but this is the one I like
fig, ax = plt.subplots()

# There are many ways to add lines
lines = ax.plot(np.array([x, x]).T, np.array([y_sin, y_cos]).T, marker=".")
assert len(lines) == 2
assert type(lines[0]) == mpl.lines.Line2D
assert lines[0].axes is ax and lines[0].figure is fig
np.testing.assert_allclose(lines[0].get_xdata(), x)

# Once we have added lines, we can modify them
lines[0].set_alpha(0.1)
lines[0].set_label("sin")
lines[1].set_label("cos")


# Legend
assert ax.get_legend() is None
ax.get_legend_handles_labels() # Note the handles are the lines and the labels are the labels!

ax.legend(numpoints=4) # note that this can be called with just the labels or both handles + labels.
# But you probably don't want to do that because you need to order things correctly.
# So, continue to label things when you plot the line. But if you only want legend for
# a subset of lines, ax.legend([l1, l2], [lab1, lab2]) is a way
```
