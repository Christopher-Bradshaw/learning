# Steepest Descent

$$
f(x_1, x_2) = (cx_1 − 2)^4 + x_2^2(cx_1 − 2)^2 + (x_2 + 1)^2
$$

```python
def get_f(c):
    return lambda x: (c*x[0] - 2)**4 + x[1]**2 * (c*x[0] - 2)**2 + (x[1] + 1)**2
```

```python

for c, lims in [
    (10, [0, 0.3, -2, 2]), 
    (1, [0, 4, -2, 2])
]:
    f = get_f(c)
    grad_f = grad(f)
    x0 = np.array([3., 3.]) # Why not

    ax = p.surface_2d(f, lim=lims, typ="contour")

    positions = algos.steepest_descent(f, grad_f, x0)
    print(len(positions))
    ax.plot(positions[:,0], positions[:,1])
```
