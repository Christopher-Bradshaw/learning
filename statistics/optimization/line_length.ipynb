# Line length

```python
f = lambda x: x ** 4 + x ** 2
x0 = np.array([2.1])
p = - grad(f)(x0)
a = algos.get_line_length(f, grad(f), x0, p, 10)

x = np.linspace(-3, 3, num=300)
_, ax = plt.subplots()
ax.plot(x, f(x))
pts = np.array([x0, x0 + p * a])
ax.scatter(pts, f(pts))
```
