# Histogram Equalization

```python
img = cv2.imread("moon.jpg")[:,:,2] # let's just deal in a single band first
print(img.shape, img.dtype)
```

```python
_, ax = plt.subplots()
ax.imshow(img, plt.get_cmap("Greys_r"))

_, ax = plt.subplots()
counts, bins, _ = ax.hist(img.flatten(), density=True, bins=np.arange(256))
cdf = np.cumsum(counts)
ax.plot(bins[1:], cdf / cdf[-1] * np.max(counts))
```
